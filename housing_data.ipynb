```python
# Importing the dependenices

import pandas as pd
from pathlib import Path

# Reading the data
housing_data =Path("Resources/BC_cities Housing_data.csv")

housing_df =pd.read_csv(housing_data)

housing_df.head()

```

```python
#Checking the age_groups how many it has and dropping the extra columns
housing_df['Age groups'].value_counts()
new_df=housing_df.drop(['SCALAR_FACTOR','SCALAR_ID','STATUS','SYMBOL','TERMINATED'],axis=1)
new_df.head()
```

```python
# Dropping null values from the data set
Clean_data=new_df.dropna(subset=['VALUE'])
Clean_data
```

```python
#Checking if the data has any null values
Clean_data["VALUE"].isnull().value_counts()
```

```python
#Checking the values for the estimate column in the data
Clean_data["Estimates"].value_counts()

```

```python
#Making age groups list to get a Dataframe only for those 3 age groups

Age = ['35 to 54 years old',"55 years old and over","Under 35 years old"]

Average_price_df =Clean_data.loc[Clean_data["Age groups"].isin(Age)]

Average_price_df

```

```python
Average_price_df["Estimates"].value_counts()
```

```python
Average_price_2018 = Average_price_df.loc[(Average_price_df["REF_DATE"]==2018)&
    (Average_price_df["Number of buyers as part of a property sale"]=="Total, all number of buyers categories")&
    (Average_price_df["Buyer characteristics"]=="Total, all admission categories")&
    (Average_price_df["First-time home buyer status"]=="Total, all home buyers categories")&
    (Average_price_df["Estimates"]=="Average sale price")
    ]

Average_price_2018
```

```python
# Average sale price for each Age group 
Average2018=Average_price_2018.loc [(Average_price_2018["Age groups"]=="35 to 54 years old")&(Average_price_2018["Estimates"]=="Average sale price")]
Average_sale= Average2018["VALUE"].mean()

#Average sale price for each Age group 
Average2018=Average_price_2018.loc [(Average_price_2018["Age groups"]=="Under 35 years old")&(Average_price_2018["Estimates"]=="Average sale price")]
Average_sale2=Average2018["VALUE"].mean()

#Average sale price for each Age group 
Average2018=Average_price_2018.loc [(Average_price_2018["Age groups"]=="55 years old and over")&(Average_price_2018["Estimates"]=="Average sale price")]
Average_sale3=Average2018["VALUE"].mean()
```

```python
Average_price_2019 = Average_price_df.loc[(Average_price_df["REF_DATE"]==2019)&
    (Average_price_df["Number of buyers as part of a property sale"]=="Total, all number of buyers categories")&
    (Average_price_df["Buyer characteristics"]=="Total, all admission categories")&
    (Average_price_df["First-time home buyer status"]=="Total, all home buyers categories")&
    (Average_price_df["Estimates"]=="Average sale price")
    ]

Average_price_2019
```

```python
# Average sale price for each Age group 
Average2019=Average_price_2019.loc [(Average_price_2019["Age groups"]=="35 to 54 years old")&(Average_price_2019["Estimates"]=="Average sale price")]
Average_sale4= Average2019["VALUE"].mean()

#Average sale price for each Age group 
Average2019=Average_price_2019.loc [(Average_price_2019["Age groups"]=="Under 35 years old")&(Average_price_2019["Estimates"]=="Average sale price")]
Average_sale5=Average2019["VALUE"].mean()

#Average sale price for each Age group 
Average2019=Average_price_2019.loc [(Average_price_2019["Age groups"]=="55 years old and over")&(Average_price_2019["Estimates"]=="Average sale price")]
Average_sale6=Average2019["VALUE"].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

Age_2018 =['35 to 54 years old',"Under 35 years old","55 years old and over"]
Average_sale_2018 =[Average_sale,Average_sale2,Average_sale3]

colors = ["blue","green","lightblue"]
explode=(0.12,0,0)
plt.pie(Average_sale_2018, labels=Age_2018, colors=colors, explode=explode,autopct="%1.1f%%")

plt.title("2018 :Average Sale Price vs Age")               
# Create axes which are equal so we have a perfect circle
plt.axis("equal")
# Save an image of our chart and print the final product to the screen

plt.show()
```

```python
#Regression line for number of buyers value vs Average buyer age 
Average_price_df
#Average buyer Sale price of first time buyers 
avg_price=Average_price_2018[(Average_price_2018["Estimates"]=="Average sale price")&(Average_price_2018["Age groups"]=="35 to 54 years old")]["VALUE"]

avg_price
```

```python
# number of buyer for first time home 
buyer_number=Average_price_df.loc[(Average_price_df["Estimates"]=="Average buyer age")&
                                  (Average_price_df["Age groups"]=="35 to 54 years old")&
                                  (Average_price_df["Number of buyers as part of a property sale"]=="Total, all number of buyers categories")&
                                  (Average_price_df["Buyer characteristics"]=="Total, all admission categories")&
                                  (Average_price_df["First-time home buyer status"]=="Total, all home buyers categories")&
                                  (Average_price_df["REF_DATE"]==2018)]["VALUE"]

buyer_number
```

```python
Comparison_df=pd.DataFrame({"avg_price":avg_price.values,"buyer_number":buyer_number.values})
```

```python
# Regression line
import scipy.stats as lineregress
from scipy import stats

(slope,intercept,rvalue,pvalue,stderr)= stats.linregress(Comparison_df["avg_price"],Comparison_df["buyer_number"])
regress_values = Comparison_df["avg_price"] * slope + intercept

# Create line 

line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

#create plot 
plt.scatter(Comparison_df["avg_price"],Comparison_df["buyer_number"])
plt.plot(avg_price,regress_values,"r-")

# labeling the data
plt.xlabel("Average Sales Price")
plt.ylabel("Average Buyer Age")
print(f"Regress value -r: {rvalue}")
plt.annotate(line_eq,(500000,43),color="blue",fontsize=15)
plt.title("Average buyer age vs Average sale price 2018")
plt.show()
```
